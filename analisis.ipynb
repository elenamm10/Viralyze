## Importaci√≥n de librer√≠as

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Definici√≥n de rutas y carga de datos

```python
lote1_path = "instagram_lote1.csv"
lote2_path = "instagram_lote2.csv"
df1 = pd.read_csv(lote1_path)
df2 = pd.read_csv(lote2_path)
df = pd.concat([df1, df2], ignore_index=True)
df.head()
```

## Informaci√≥n general de la BBDD

```python
df.info()

# estad√≠sticas b√°sicas para variables num√©ricas
df.describe()
```

## Limpieza de los datos

```python
#  Modificar tipos de variables:
    # - hour_of_day, day_of_week, media_type -> DUMMY

for hour in df['hour_of_day'].unique():
    df[f'hour_of_day_{hour}'] = df['hour_of_day'].apply(lambda x: 1 if x == hour else 0)

for day in df['day_of_week'].unique():
    df[f'day_of_week_{day}'] = df['day_of_week'].apply(lambda x: 1 if x == day else 0)

for media in df['media_type'].unique():
    df[f'media_type_{media}'] = df['media_type'].apply(lambda x: 1 if x == media else 0)
```

```python
df.head()
```

## An√°lisis univariado de las variables categ√≥ricas

```python
cat_cols = ['hour_of_day', 'day_of_week', 'media_type']
for col in cat_cols:
    sns.countplot(data=df, x=col)
    plt.title(f"Distribuci√≥n de {col}")
    plt.xticks(rotation=45)
    plt.show()
```

#### Distribuci√≥n de la hora del d√≠a
El histograma muestra que las publicaciones se concentran mayoritariamente entre las 11:00 y las 14:00, con un pico destacado a las 19:00. Este √∫ltimo horario es especialmente relevante, ya que coincide con una franja de baja actividad en redes sociales (modelo.ipynb). En contraste, las horas de madrugada y primeras horas del d√≠a (antes de las 8:00) tienen muy baja frecuencia de publicaci√≥n.


#### Distribuci√≥n del d√≠a de la semana
La mayor√≠a de publicaciones se concentran entre lunes y viernes. Los fines de semana (especialmente s√°bado) tienen una participaci√≥n considerablemente menor, lo que podr√≠a indicar una menor actividad por parte de los creadores de contenido o un menor rendimiento percibido. Esta tendencia sugiere que, estrat√©gicamente, publicar entre semana puede alinearse mejor con el comportamiento de los usuarios y del algoritmo (? Lo comprobaremos).


#### Distribuci√≥n del tipo de publicaci√≥n
Los tres formatos de contenido (reel, carrusel y foto) est√°n equilibradamente representados en el conjunto de datos.


## An√°lisis univariado de las variables num√©ricas

```python
num_cols = ['likes', 'comments', 'caption_length', 'num_hashtags', 'engagement_ratio', 'num_mentions', 'num_emojis'] 

for col in num_cols:
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribuci√≥n de {col}")
    plt.show()
```

#### Distribuci√≥n de likes (`likes`)

Los likes tienen una distribuci√≥n fuertemente sesgada a la derecha (distribuci√≥n de cola larga). La mayor√≠a de publicaciones reciben pocos likes, mientras unas pocas reciben miles. Esto sugiere que hay outliers muy virales que podr√≠an estar influyendo mucho en el aprendizaje del modelo.

#### Distribuci√≥n de comentarios (`comments`)

Similar a los likes, la mayor√≠a de publicaciones recibe pocos comentarios. Solo unas pocas generan una cantidad muy alta, indicando que el engagement elevado por comentarios es menos frecuente y m√°s dif√≠cil de lograr.

#### Longitud del caption (`caption_length`)

La mayor√≠a de captions tienen entre 100 y 600 caracteres, con un pico alrededor de 300. Captions m√°s largos no necesariamente implican m√°s interacci√≥n (lo veremos), aunque aportan m√°s contexto. Hay publicaciones con captions extremadamente largos (outliers), pero son poco comunes.

#### N√∫mero de hashtags (`num_hashtags`)

La mayor√≠a de publicaciones usa entre 0 y 10 hashtags, con un valor modal en 2-3 hashtags. Esto sugiere que los creadores tienden a ser moderados en su uso.

#### N√∫mero de menciones (`num_mentions`)

La mayor√≠a de publicaciones no incluyen ninguna menci√≥n. Solo unas pocas publicaciones mencionan a varias cuentas.

#### N√∫mero de emojis (`num_emojis`)

Los usuarios tienden a usar entre 1 y 4 emojis por publicaci√≥n. El uso de emojis est√° muy correlacionado con la emocionalidad y expresividad del texto, lo que puede tener efecto positivo sobre la viralidad si se usa adecuadamente (tambi√©n lo comprobaremos).

#### Ratio de engagement (`engagement_ratio`)

El ratio de engagement muestra una distribuci√≥n extremadamente sesgada, con la mayor√≠a de valores cerca de cero y unos pocos muy altos. Esto indica que la mayor√≠a de publicaciones tienen poco engagement en relaci√≥n a su base de seguidores, y unas pocas generan gran impacto relativo, estas son las que nos interesan.


## Boxplot de `engagement_ratio` seg√∫n el d√≠a de la semana

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="day_of_week", y="engagement_ratio",
            order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
plt.title("Engagement Ratio seg√∫n D√≠a de la Semana")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

El gr√°fico boxplot muestra la variabilidad del `engagement_ratio` a lo largo de los distintos d√≠as de la semana:

- **Saturday (s√°bado)** destaca claramente como el d√≠a con mayor mediana de engagement y con una distribuci√≥n mucho m√°s amplia. Esto sugiere que los s√°bados pueden ser especialmente eficaces para lograr interacciones, con algunos posts alcanzando niveles muy altos.
- **Friday (viernes)** tambi√©n muestra cierta elevaci√≥n en engagement, aunque con menor dispersi√≥n que s√°bado.
- Los dem√°s d√≠as (**lunes a jueves y domingo**) presentan distribuciones m√°s bajas y concentradas en torno a valores reducidos, lo que sugiere menor eficacia media en t√©rminos de engagement.

## Boxplot de `engagement_ratio` seg√∫n la hora del d√≠a

```python
plt.figure(figsize=(12, 5))
sns.boxplot(data=df, x="hour_of_day", y="engagement_ratio")
plt.title("Engagement Ratio seg√∫n Hora del D√≠a")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Este boxplot revela c√≥mo var√≠a el `engagement_ratio` a lo largo de las distintas horas del d√≠a:

- **Primeras horas de la ma√±ana (5 a 7)** presentan una mediana de engagement significativamente alta, as√≠ como una mayor dispersi√≥n. Esto sugiere que publicar en esas horas podr√≠a maximizar la interacci√≥n, probablemente debido a menos saturaci√≥n de contenido.
- **Horas centrales (12 a 13)** tambi√©n muestran valores altos de engagement en algunos casos, pero con mayor variabilidad y presencia de outliers.
- **Tarde-noche (17 en adelante)** tiene una ca√≠da clara en la mediana y una concentraci√≥n de valores bajos, lo que indica menor eficacia en t√©rminos de engagement.

## Boxplot de `engagement_ratio` seg√∫n el tipo de contenido

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="media_type", y="engagement_ratio")
plt.title("Engagement Ratio seg√∫n Tipo de Publicaci√≥n")
plt.tight_layout()
plt.show()
```

Este gr√°fico analiza c√≥mo var√≠a el `engagement_ratio` en funci√≥n del formato de publicaci√≥n:

- **Fotos** presentan una mediana de engagement claramente m√°s alta que reels o carruseles, y una distribuci√≥n m√°s amplia con numerosos valores at√≠picos. Esto indica que, aunque no siempre, las fotos pueden alcanzar ratios de interacci√≥n muy elevados.
- **Reels** y **carruseles** muestran medianas mucho m√°s bajas y menos dispersi√≥n, lo que sugiere una interacci√≥n m√°s contenida o estable, aunque tambi√©n aparecen outliers con buen rendimiento.

Aunque com√∫nmente se espera que los reels lideren en engagement, en este conjunto de datos las **fotos** han demostrado un mayor potencial viral. Considera priorizar fotos si el contenido lo permite, pero sigue combinando formatos para maximizar alcance y variedad.


## Matriz de correlaci√≥n entre variables num√©ricas

```python
# Crear variable objetivo binaria (viral si supera el Q3 de engagement)
q75 = df["engagement_ratio"].quantile(0.75)
df["is_viral"] = df["engagement_ratio"] > q75
display(df[["engagement_ratio", "is_viral"]].head())
```

```python
variables = ['likes', 'comments', 'caption_length', 'num_hashtags', 'engagement_ratio', 'num_mentions', 'num_emojis', 'is_viral']
corr = df[variables].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlaci√≥n")
plt.show()
```

### Top variables correlacionadas con nuestra variable objetivo

| Variable             | Correlaci√≥n con `is_viral` | Interpretaci√≥n |
|----------------------|-----------------------------|----------------|
| `likes`              | **0.59**                    | Fuerte relaci√≥n positiva. Obvio porque para calcular is_viral hemos usado engagement_ratio que usa likes. |
| `engagement_ratio`   | **0.57**                    | Muy correlacionado con la viralidad. Idem que like. |
| `num_hashtags`       | **0.55**                    | Relaci√≥n positiva clara. Usar hashtags adecuados y en cantidad razonable aumenta la visibilidad y, por tanto, la viralidad. |
